```python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import pandas as pd
```

PCA : Combina os atributos


```python
dataset = pd.read_csv(r'C:\Users\Edno\Desktop\Work\ALANA\PPSO-master\ransomware.csv')
```

```python
dataset
```

```python
train_row, teste_row = 600,600
columns = 50

tipo = dataset.iloc[:train_row, 1]
dados = dataset.iloc[:train_row, 3:columns+3]

tipo_teste = dataset.iloc[teste_row:, 1].reset_index(drop=True)
dados_teste = dataset.iloc[teste_row:, 3:columns+3].reset_index(drop=True)
```

```python
pca = PCA(n_components=15)
```

```python
X_treinamento_pca = pca.fit_transform(dados.values)
X_testes_pca = pca.transform(dados_teste.values)
```

```python
X_treinamento_pca.shape, X_testes_pca.shape
```

```python
#Explicação das variáveis principais
pca.explained_variance_ratio_
pca.explained_variance_ratio_.sum()
```

```python
X_treinamento_pca = pd.DataFrame(X_treinamento_pca)
X_testes_pca = pd.DataFrame(X_testes_pca)
```

```python
X_treinamento_pca
```

```python
lda = LinearDiscriminantAnalysis(n_components = 8)
```

```python
X_treinamento_lda = lda.fit_transform(dados.values, tipo.values)
X_testes_lda = lda.transform(dados_teste.values)
```
